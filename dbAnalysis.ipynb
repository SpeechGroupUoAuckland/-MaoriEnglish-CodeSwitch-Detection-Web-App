```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

db = '20220321_Hansard_DB_all.csv'
db_MP_only = '20220321_Hansard_DB_MP_only.csv'
db_MP_only_window_3 = '20220321_Hansard_DB_MP_only_window_3.csv'
```

```python
pdDb = pd.read_csv(db)
pdDb_MP_only = pd.read_csv(db_MP_only)
pdDb_MP_only_window_3 = pd.read_csv(db_MP_only_window_3)

pdDb['text'] = pdDb['text'].astype(str)
pdDb_MP_only['text'] = pdDb_MP_only['text'].astype(str)
pdDb_MP_only_window_3['text'] = pdDb_MP_only_window_3['text'].astype(str)
```

```python
def calcNumberOfWords(text: str) -> int:
    '''Use label as input'''
    return text.count(',')+1
```

### Original data


```python
# length of the database
print('Length of the database:', len(pdDb))
# count the number of different labels in the database
print('Number of labels: \n\t' + str(pdDb['Labels_Final'].value_counts()).replace('\n', '\n\t')[:-34])
print()


# calculate number of words use label column
y = pdDb['label'].apply(calcNumberOfWords).to_list()

# plot y, add x-axis the index of y, y-axis y
plt.plot(y, '.')
plt.xlabel('Index')
plt.ylabel('Number of words')
plt.show()

# print the min, max, mean, and std of the number of words with their index in the database
print('Number of words: \n\tMin:', min(y), ':', y.index(min(y)), \
                        '\n\tMax:', max(y), ':', y.index(max(y)), \
                        '\n\tMean:', np.mean(y), \
                        '\n\tStd:', np.std(y))


# calculate the length of text in the database
y = pdDb['text'].apply(len).to_list()

# plot y, add x-axis the index of y, y-axis y
plt.plot(y, '.')
plt.xlabel('Index')
plt.ylabel('Length of text')
plt.show()

# print the min, max, mean, and std of the length of text with their index in the database
print('Length of words: \n\tMin:', min(y), ':', y.index(min(y)), \
                        '\n\tMax:', max(y), ':', y.index(max(y)), \
                        '\n\tMean:', np.mean(y), \
                        '\n\tStd:', np.std(y))
```
